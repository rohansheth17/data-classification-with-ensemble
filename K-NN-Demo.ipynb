```python
traindata = [line.strip().split('\t') for line in open('project3_dataset3_train.txt', 'r')]
```

```python
import numpy as np
traindata = np.asarray(traindata)
print(traindata)
classlabels = traindata[:,-1].reshape((traindata.shape[0],1)).astype(int)
traindata = traindata[:,0:-1]
print(classlabels)
print(traindata)
```

```python
testdata = [line.strip().split('\t') for line in open('project3_dataset3_test.txt', 'r')]
testdata = np.asarray(testdata)
print(testdata)
testlabels = testdata[:,-1].reshape((testdata.shape[0],1)).astype(int)
testdata = testdata[:,0:-1]
print(testlabels)
print(testdata)
```

```python
traindata = traindata.astype(float)
traindata = np.append(traindata,classlabels,axis=1)
print(traindata)
testdata = testdata.astype(float)
testdata = np.append(testdata,testlabels,axis=1)
print(testdata)
```

```python
print(traindata.shape)
print(testdata.shape)
```

```python
def calculate_dist(p1,p2):
    sum = 0
    for i in range(len(p2)):
        sum += ((p1[i] - p2[i]) ** 2)
    dist = sum**0.5        
    return dist


def classify(train_data, point_to_classify):
    dist_list = []
    for tr in train_data:
        dist_list.append(calculate_dist(tr[0:-1],point_to_classify[0:-1]))
    closestk = np.argsort(np.asarray(dist_list))[:k]
    zeros = 0
    ones = 0
#     print(len(closestk))
    for c in closestk:
#         print(c)
        if(train_data[:,-1][c] == 0):
            zeros +=1
        elif(train_data[:,-1][c] == 1):
            ones +=1
#     print("------------------------------------------------------------------------")        
    return 1 if(max(ones,zeros) == ones) else 0
    
    
    
```

```python
def evaluate(actual, predicted):
    # a =TP, b=FN, c= FP, d= TN
    a = b = c = d = 0
    for i in range(len(actual)):
        if(actual[i] == 1 and predicted[i] == 1):
            a += 1
        elif(actual[i] == 1 and predicted[i] == 0):
            b += 1
        elif(actual[i] == 0 and predicted[i] == 1):
            c += 1
        elif(actual[i] == 0 and predicted[i] == 0):
            d += 1
    return a, b, c, d
```

```python
accuracy = precision = recall = f_measure = 0
k=6
train, test = traindata, testdata
test_labels = []
for j in range(len(test)):
    test_labels.append(classify(train, test[j]))
    
print(len(test_labels)) 
test_labels = np.asarray(test_labels)
a, b, c, d = evaluate(test[:,-1], test_labels)

accuracy += (float(a + d)/(a + b + c + d))
precision += (float(a)/(a + c))
recall += (float(a)/(a + b))
f_measure += (float(2 * a) / ((2 * a) + b + c))
    
print("Accuracy: ",accuracy*100,"%")
print("Precision: ",precision*100,"%")
print("Recall: ",recall*100,"%")
print("F-1 Measure: ",f_measure*100,"%")   
```
